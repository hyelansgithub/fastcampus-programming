#### Thread

- 프로그램이 동작 될 때 실행시간이 오래 걸린다면 두가지 문제가 있기 때문입니다.
- I/O 바운드 : 파일을 읽고 쓰거나, 네트워크 시간등의 문제
- CPU 바운드 : 계산량이 많은 경우 (코드의 효율화나 알고리즘을 통해서 해결해야 한다.)
- I/O 바운드의 문제를 thread 를 통해서 어느정도 해결할 수 있습니다.

- 동기 sync
    - 작업이 끝나야 다음 작업이 수행됩니다.
- 비동기 async
    - 작업이 독립적으로 수행됩니다.

```python
import threading # 모듈 임포트
```

```python
ls1, ls2 = [], []
```

```python
def th_func1(number):
    for idx in range(number):
        ls1.append(idx)
        if idx % 100 == 0:
            print("ls1 :", idx)

def th_func2(number):
    for idx in range(number):
        ls2.append(idx)
        if idx % 100 == 0:
            print("ls2 :", idx)
            
```

```python
th1 = threading.Thread(target = th_func1, args=(12000,)) 

th2 = threading.Thread(target = th_func2, args=(10000,)) 
th1.start()
th2.start()
```

```python
#thread1 에 A페이지 넣고
#thread2 에 B페이지 넣어서 동시에 크롤링 할 수 있다.
```
