##### 아래 URL의 NBA 데이터를 크롤링하여 판다스 데이터 프레임으로 나타내세요. 
- http://stats.nba.com/teams/traditional/?sort=GP&dir=-1

```python
from selenium import webdriver
```

```python
import pandas as pd
```

```python
def Nba():

    try:

        driver = webdriver.Chrome()
        driver.get("http://stats.nba.com/teams/traditional/?sort=GP&dir=-1")

        # column_list
        columns = driver.find_elements_by_css_selector(".nba-stat-table__overflow > table > \
        thead > tr > th")[1:28]
        column_list = [" ",]
        for column in columns:
            column = column.text
            column_list.append(column)

        # making dataframe
        df = pd.DataFrame(columns=column_list, index = [" " for i in range(30)])

        # data
        data = driver.find_elements_by_css_selector\
        (".nba-stat-table__overflow > table > tbody > tr > td")
        data_list = []

        for datum in data:
            datum = datum.text.replace("\n", " ")
            data_list.append(datum)

        grouped = list(zip(*[iter(data_list)] * 28))

        for i in range(30):
            for j in range(28):
                df.iloc[i, j] = grouped[i][j]

        driver.quit()

    except:

        driver.quit()

    return df
```

```python
Nba()
```

- http://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=105

```python
def News():
    
    try:

        driver = webdriver.Chrome()
        driver.get("http://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=105")

    #     news = pd.DataFrame(columns = ["Number of Articles","Article Headline"])

        for i in range(10):
            results = driver.find_elements_by_css_selector("#section_body > ul > li > dl > dt:nth-child(2) > a")
            for result in results:
                result = result.text
                print(result)
            
            driver.find_element_by_css_selector("#paging > a:nth-child({})".format (i+2)).click()

        driver.quit()
    
    except:
        driver.quit()
```
