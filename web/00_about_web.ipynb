#### 1. Server & Client Architecture

    a. 검색어 입력 : client가 브라우저의 검색 엔진 웹페이지 검색창에 '데이터과학'이라는 검색어를 입력하며 서버로 Request 요청, 요청한 검색어가 라우터로 가고, 이어서 인터넷을 통해 검색엔진 회사의 데이터 센터로 간다.

    b. 서버에서 목록 검색 : 데이터 센터의 검색 엔진 컴퓨터는 '데이터', '과학'이라는 단어를 찾으며 웹페이지 목록을 훑는다. 검색 결과를 담은 웹 페이지를 인터넷을 통해 client 컴퓨터로 회신한다. 각 결과에는 URL링크가 달려있다. 이때 브라우저가 HTML, CSS, JS 코드를 분석 및 해석한 후 검색 결과 화면을 띄워준다.
    
    d. 검색 결과에서 원하는 웹페이지 요청 : 이후 사용자가 페이지의 URL을 클릭하면 새로운 request를 라우터를 통해 인터넷을 거쳐 웹 페이지가 저장된 인터넷 서비스 제공자(ISP)서버로 전달한다.

    e. 웹페이지 보내기 : 검색엔진이 찾은 결과 중 사용자가 찾는 검색어에 맞는 것이 하나있는데 웹서버는 이 페이지와 이 페이지에 있는 영상을 인터넷을 통해 사용자 컴퓨터로 보낸다.
    
    
#### 2. URL(Uniform Resource Locator)

URL provides a way to locate a resource on the web, the hypertext system that operates over the internet.

예시 : `http://` `news` `.naver.com` `:80` `/main/` `read.nhn`

`?mode=LSD&mid=shm&sid1=105&oid=001&aid=0009847211` `#da_727145`


- `Protocol`: 네트워크 상에서 다른 컴퓨터끼리 정보를 주고받기 위한 통신 규약 https://, http://

- `Sub Domain` : sub domain으로 서비스를 구분한다. news

- `Domain` : 네트워크 상의 컴퓨터 위치 naver.com :

- `port` : 서버로 들어가기 위한 문, 열려있지 않으면 접속 불가하다. 80
    - port가 localhost일때 : 내 컴퓨터로 접속한다는 의미

- `path` : 서버 내의 디렉토리 /main/

- `page` : read.nhn

- `query` : ?mode=LSD&mid=shm&sid1=105&oid=001&aid=0009847211
    - ? : 페이지와 쿼리를 나눠주는 역할을 한다.

- `fragment` : 가져온 페이지의 위치를 나타낸다. #da_727145



#### 3. Get & Post

- get : url에 데이터가 포함, 데이터가 노출되는 장단점 가짐, 길이 제한이 있음, 브라우저마다 길이 제한이 다르다.

- post : body에 데이터가 포함된다. 데이터가 숨겨지고 데이터 많이 보낼 수 있다.


#### 4. Internet

- 인터넷은 컴퓨터로 연결하여 TCP/IP (Transmission Control Protocol/Internet Protocol)라는 통신프로토콜을 이용해 정보를 주고받는 컴퓨터 네트워크

- TCP/IP라는 통신 프로토콜(약속)을 사용한다.

- 해저망 : 해저케이블을 이용하여 전세계 서버에 접속하여 인터넷을 사용하고 있다.

- 무선인터넷
    - 선이 아니라 주파수를 매체로 사용한다.
    - 우리가 사용하는 단말기만 무선이다.
    
- wifi : 10미터까지 가능
    
- cellular : cell 어떤 영역을 커버할때 cell 단위로 커버한다. 셀의 크기가 도시와 지방이 다르다. 지방에는 cell 크기가 더 커서 전파를 더 멀리 쏴야해서 배터리가 빨리 닳게 되는 것이다.

    - 핸드오프 : cell범위 바꾸는 것을 뜻함
    
- 주파수별로 도달 범위가 다르다. 주파수 낮은 것은 멀리갈 수 있다. 
- 낮은 주파수 쓰면 0,1 데이터 많이 담을 수 없다.
- 유선은 외부 간섭을 적게 받아서 전송률이 높다.


#### 5. OSI 7 Layer

- Open Systems Interconnection Reference Model

- 국제표준화기구(ISO)에서 개발한 모델로, 컴퓨터 네트워크 프로토콜 디자인과 통신을 계층으로 나누어 설명한 것이다.

- 하위 계층으로 갈수록 페이로드가 증가


#### 6. Cookie & Session & Cache

- Cookie
    - Client에 저장하는 문자열 데이터로 도메인 별로 따로 저장
    - 로그인 정보, 내가 봤던 상품 정보, 팝업 다시보지 않음
    - 하나의 클라이언트에 300개, 도메인당 20개, 쿠키 하나당 4Kbyte
- Session
    - Server에 저장하는 객체 데이터, 브라우져와 연결시 Session ID 생성
    - Session ID를 Cookie에 저장함으로 로그인 연결 유지
    - 같은 브라우져로 같은 서버에 접속하면 Session ID가 같음
    - 로그인 연결 정보, 원하는 객체 데이터
- Cache
    - Client나 Server의 메모리에 저장하여 빠르게 데이터를 가져오는 목적의 저장소

#### 7. HTTP Status Code

- 서버와 클라이언트가 데이터를 주고 받으면 주고 받은 결과로 상태 코드를 확인할수 있다.
    - 2xx - success
    - 3xx - redirection (browser cache)
    - 4xx - request error
    - 5xx - server error
    - http://bit.ly/2nlZM8L

#### 9. Scraping & Crawling & Spider & Bot 

- Scraping : 데이터를 수집하는 작업
- Crawling : 여러페이지의 특정 데이터들을 수집하고 분류하는 작업 (크롤링이 조금 더 큰 개념이고 해외에서는 Scraping 더 많이 쓴다.)
- Spider or Web crawler : 웹 데이터를 수집하는 소프트웨어
- Bot : 인터넷 상에서 자동화된 작업을 실행하는 소프트웨어 (봇이 스파이더보다 조금 더 발전된 형태)


- `/robots.txt`: 데이터 수집 허용범위 알 수 있다. https://www.naver.com/robots.txt
